```python
def calculate_average_income(dataframe):
    columns_to_convert = ['Less than $10,000',
       '$10,000 to $14,999', '$15,000 to $24,999', '$25,000 to $34,999',
       '$35,000 to $49,999', '$50,000 to $74,999', '$75,000 to $99,999',
       '$100,000 to $149,999', '$150,000 to $199,999', '$200,000 or more']
    
    for column in columns_to_convert:
        dataframe[column] = pd.to_numeric(dataframe[column], errors='coerce')  # Convert to numeric (float or NaN)
        
    mask = dataframe[columns_to_convert].apply(lambda x: x == '-', axis=0).any(axis=1)
    dataframe = dataframe[~mask]  # Remove rows with '-' values in selected columns
        
    dataframe['avg_household_income'] = (dataframe['Less than $10,000'] * 5000 +
                                   dataframe['$10,000 to $14,999'] * 12500 +
                                   dataframe['$15,000 to $24,999'] * 20000 +
                                   dataframe['$25,000 to $34,999'] * 30000 +
                                   dataframe['$35,000 to $49,999'] * 42500 +
                                   dataframe['$50,000 to $74,999'] * 62500 +
                                   dataframe['$75,000 to $99,999'] * 87500 +
                                   dataframe['$100,000 to $149,999'] * 125000 +
                                   dataframe['$150,000 to $199,999'] * 175000 +
                                   dataframe['$200,000 or more'] * 250000) / 100
    return dataframe
```

```python

 # Function to check if a violation description is contained in the dictionary values
def get_violation_label(description):
    for label, descriptions in lowercase_dict.items():
        if any(desc in description for desc in descriptions):
            return label
    return None
 # Apply the function to create the 'ViolationLabel' column
merged_income_dem2['ViolationLabel'] = merged_income_dem2['ViolationDescription'].apply(lambda x: get_violation_label(x))
```
